## <font color="blue">场景文本导出的运用场景如下：</font>

- 对某些**含有大量场景对话或内置脚本**的自定义场景进行文本内容修改与维护时
- 需要对国外作者的**场景文本进行翻译**时

以上场合均涉及到场景文本编辑，如果在地图编辑器中查找编辑，将会相当麻烦。那么，有没有一种办法将场景中的文本导出，编辑好之后再导入呢？答案是肯定的，可以利用ASP库进行文本导出！

## <font color="blue">利用ASP导出场景文本 </font>
    在前几个部分的文档中，我们已经了解了ASP的一些使用功能在本部分，将介绍ASP导出场景中的文本。

### <font color="orange">关于场景文本提取脚本的一些说明：</font>

1.此脚本使用ASP库加载场景中必要的场景对话和游戏脚本，将其导出为 CSV文件  
2.玩家只需设置自己的AOE2DE游戏ID，需要导出的场景名称（默认位于 C:/Users/<USERNAME>/Games/Age of Empires 2 DE/[game_id]/resources/_common/scenario），即可将文本进行导出

关于此脚本的详细使用说明，请参考此脚本的 **introduce** 部分。

# <font color="blue">脚本使用方法：</font>

## 关于文本提取的共功能已经封装到Python脚本 <font color="purple">"P4-场景文本提取.py"</font> 中。

### 1.在终端执行以下命令之一运行脚本<font color="purple">"P4-场景文本提取.py"</font>，导出场景文本

```python
# 执行无参数脚本，以预设信息寻找对应场景进行提取
    python P4-场景文本提取.py
```

```python
# 执行有参数脚本，通过终端参数寻找对应场景进行提取（文件后跟的2个参数分别代表： 游戏ID、场景名称）
    python P4-场景文本提取.py 76561199072658060 Sanatorium
```

### 2.参照 introduce 脚本说明对导出的文本进行修改

````python
# 脚本使用说明
introduce = \
"""本程序运行后，执行以下操作：
1. 读取场景文件中所有的文本，输出到指定的文件中
2. 创建新的场景，将翻译后的文本填入场景中
    请注意，这个程序只能翻译不使用字符串表 ID 的场景，如果场景中使用了字符串表，
请直接新建 ./resources/zh/strings/key-value/key-value-modded-strings-utf8.txt 文件翻译，
这样的翻译文本在游戏语言为简体中文时会自动被使用。本程序依然会导出含有字符串表 ID 的原始字符串及其对应的 ID，
因为字符串表 ID 对应字符串不存在时，游戏将会使用回退文本（仅部分触发情况如此），
译者也可以根据这些 ID 书写 key-value-modded-strings-utf8.txt 文件。
    导出文件为 csv 格式，若不熟悉该格式的实现细节，请使用 Excel 或同类电子表格软件编辑，
请不要直接用 Excel 打开，因为 Excel 会将行中特殊的字符解释为公式，打开方式是：
1. 新建一个空白的工作簿
2. 选择标题栏中的 “数据” -> 获取和转换数据 -> 从文本/CSV，这一步也可以用快捷键 Alt + A + F + T   
3. 在打开的对话框中，选择本程序输出的文件导入
4. Excel 会自动读取文件编码和分隔符等细节，只需在弹出的对话框中点击“加载”
5. 加载完成后，单击表格左上角以全选表格（或者键入 Ctrl + A），
   然后选择标题栏中的 “开始” -> 数字 -> 数字格式（或者键入 Ctrl + 1），
   将单元格格式分类从 “常规” 改成 “文本”
    通过这种方式打开的表格，无需注意任何格式的问题（除了 <> 方括号里的内容不需要翻译外），
因 Excel 自动完成了这些琐碎的细节转换。如若直接编辑，请保存为带 BOM 的 UTF-8 格式，分隔符默认逗号，特殊字符用引号包含。
    请直接在译文一栏中填写译文，在 Excel 中，可通过快捷键 Ctrl + Shift + U 展开多行编辑栏。
文件对行顺序不敏感，可任意调换行顺序，但请整行调换。内部通过第一列 “类型” 来识别本行内容，
对于译文表格，必须保留 “类型” 和 “译文” 两列的内容，且第一栏标题栏不能做修改。
    可以仅修改源代码中 TODO 标注的路径，然后直接运行两次，第一次是从原始场景中提取原文，第二次是翻译好后用译文编译场景。
请注意两次运行期间，场景不能有任何改动。虽然可以先运行一次然后翻译润色一个月再运行第二次，但期间场景不能有任何改动，
如果是订阅的场景，考虑复制到个人数据路径下的场景文件夹，防止因模组作者更新后因场景改动导致无法正确导入译文。
"""
````

### 3.保存修改后的文档，再次运行脚本，更新的文本信息将导入到场景中。
